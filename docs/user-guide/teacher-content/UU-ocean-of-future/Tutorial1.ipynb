# Virtual Ship exercise

## Form and register your duo (preferably with different backgrounds) to choose a year.
- Instructions Appy

## Check out your expeditions measurement locations
https://nioz.marinefacilitiesplanning.com/cruiselocationplanning#
- Save and upload the cruise track (xls file)
- Check out the measurement locations and depths, and travel time between stations

## Choose a week within the summer half year (April - October) for your expedition
- Consider weather conditions and daylight hours for optimal data collection

## Open a terminal and prepare your expedition in virtual ship:
```virtualship plan GROUP_YOURGROUPNUMBER```
- Replace YOURGROUPNUMBER with your actual group number (e.g., GROUP_1997)

## Follow the instructions in the terminal to set up your expedition
- From the Schedule Editor select Waypoints & Instrument Selection
- Click each waypoint to set the time (consider _transit times_ between locations and _time needed for measurements_) 
- Select the instruments to use at each location: i.e. CTD and CTD-BGC
- Save your changes

## Fetch the data needed for your expedition
`virtualship fetch GROUP_YOURGROUPNUMBER --username <USERNAME> --password <PASSWORD>`
- Replace EXPEDITION_NAME with your actual expedition name (e.g., GROUP_1997)
- Provide your Copernicus username and password when prompted
- (Sign up for a Copernicus account if you don't have one yet: https://data.marine.copernicus.eu/register)
- Practice your patience - a key skill in oceanography! - data download may take some time.

## Start your expedition
`virtualship run GROUP_YOURGROUPNUMBER`

## Verify your expedition was successful by checking the output directory for data files
- Navigate to the results directory:
`cd GROUP_YOURGROUPNUMBER/results`
- Send the link to your results in via Brightspace before Wednesday 19-11-2025 13:00h.
`pwd`
